```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import logit

%matplotlib inline
```

```python
alpha = 0.05
vals = [[i, logit(alpha + (1.-alpha) * i / 256.)] for i in range(256)]
df = pd.DataFrame(vals, columns=['raw pixel', 'logit(alpha + (1-alpha)*x/256)'])
df
```

```python
df.hist(bins=32, figsize=(8, 3))
```
